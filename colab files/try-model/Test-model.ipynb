```python
import torch
from transformers import AutoModel, AutoTokenizer
from transformers import TextStreamer
from unsloth.models import FastLanguageModel
from google.colab import drive
import os
```

```python
drive.mount('/content/drive')
```

```python
#loading tokenizer and model CUSTOM

max_seq_length = 512
dtype = None

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

if baseline_model:

    my_model = "unsloth/llama-3-8b-Instruct-bnb-4bit"

    model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = my_model,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True
    )

else: #custom model

    tokenizer = AutoTokenizer.from_pretrained(model_dir)

    model = FastLanguageModel.from_pretrained(
        model_dir,
        max_seq_length = max_seq_length,
        dtype = dtype,
        load_in_4bit=True,  # Enable 4-bit quantization
        device_map="auto",  # Automatically map the model to available devices
    )

if isinstance(model, tuple):
    model = model[0]

FastLanguageModel.for_inference(model)

chat_template = """
USER: {INPUT}
ASSISTANT: {OUTPUT}"""
```

```python
messages = [
    {"role": "user", "content": "Quali sunnu i culuri primari?"},
]
#Quali sunnu li principali citati siciliani?
#"Quali sunnu i culuri primari?"
#'Mbare come è 'a tua iurnata?
#amicu, comu ti va 'a jurnata?
#mi putissi cuntari a storia dû pinocchiu?
#Cu je Dante Alighieri? scrivi 'na biografia nica.
input_ids = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True,
    return_tensors = "pt",
).to("cuda")

text_streamer = TextStreamer(tokenizer, skip_prompt = True)
_ = model.generate(input_ids, streamer = text_streamer, max_new_tokens = 512, pad_token_id = tokenizer.eos_token_id,temperature=0.75,repetition_penalty=1.1,top_k=60,top_p=0.95,length_penalty=1.1)
```
